### Dicionário de dados<a name="dicionario"></a>

| Variável              | Descrição                                                                                                  | Tipo             |
| --------------------- |:----------------------------------------------------------------------------------------------------------:| ----------------:|
| data_ref              | Data de referência de coleta das variáveis                                                                 | object           |

#### Carregando os dados
O comando pd.read_csv é um comando da biblioteca pandas (pd.) e carrega os dados do arquivo csv indicado para um objeto *dataframe* do pandas.

```python
renda = pd.read_csv('./input/previsao_de_renda.csv')
renda.info()
renda.head(1)
```

```python
renda.drop(columns=['Unnamed: 0', 'id_cliente'], inplace=True)

print('Quantidade total de linhas:', len(renda), '\n')

print('Quantidade de linhas duplicadas:', renda.duplicated().sum(), '\n')

print('Quantidade após remoção das linhas duplicadas:', 
      len(renda.drop_duplicates()), '\n')

renda.drop_duplicates(inplace=True, ignore_index=True)
renda.info()
```

#### Entendimento dos dados - Univariada

### Pandas Profiling - Relatório interativo para análise exploratória de dados

### Entendimento dos dados - Bivariadas




### Matriz de correlação


```python
renda_cor = renda[['posse_de_imovel','qtd_filhos','idade','tempo_emprego','qt_pessoas_residencia','renda']]
(renda_cor
 .corr()
 .tail(n=1)
)
```

A partir da matriz de correlação, é possível observar que a variável que apresenta maior relação com a varíavel renda é tempo_emprego, com um índice de correlação de 38,5%.

### Matriz de dispersão

```python
sns.pairplot(data=renda,
             hue='tipo_renda',
             vars=['qtd_filhos',
                   'idade',
                   'tempo_emprego',
                   'qt_pessoas_residencia',
                   'renda'],
             diag_kind='hist')
plt.show()
```

Ao analisar o pairplot, que consiste na matriz de dispersão, é possível identificar alguns outliers na variável renda, os quais podem afetar o resultado da análise de tendência, apesar de ocorrerem com baixa frequência. Além disso, é observada uma baixa correlação entre praticamente todas as variáveis quantitativas, reforçando os resultados obtidos na matriz de correlação.

```python
cmap = sns.diverging_pallete(h_neg=100,
                             h_pos=359,
                             as_cmap=True,
                             sep=1,
                             center='light')

ax = sns.clustermap(data=renda.corr(),
                   figsize=(10,10),
                   center=0,
                   cmap=cmap)
plt.setp(ax.ax_heatmap.get_xticklaberls(), rotation=45)

plt.show()
```
